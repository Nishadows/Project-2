```python
years = ["2011","2012","2013",
                    "2014","2015","2016","2017","2018"]
print(years)
```

```python
import requests
import json


for year in years:
    print(year)
#     query_url = "https://raw.githubusercontent.com/emreynolds9/Project-2/master/Data_cleaning/va_counties.geojson"
#     response = requests.get(query_url).json()
    with open("keeping_counties.geojson") as json_file:
        data = json.load(json_file)
    for countyelem in data["features"]:
            county=countyelem["properties"]["County"]
            state = countyelem["properties"]["State"]
#             print(county)
            with open('rental.json') as json_file:
                rental_data = json.load(json_file)
                for d in rental_data:
                    if ((d["County"]==county)&(d["State"]==state)):
                        countyelem["properties"]["RENTAL_PRICE"]=d[year]
#                         print(d[year])
            with open('pricespersqft.json') as json_file:
                value_data = json.load(json_file)
                for d in value_data:
                    if ((d["County"]==county)&(d["State"]==state)):
                        countyelem["properties"]["HOME_VALUE"] = d[year]
    with open(f'../Resources/{year}.geojson', 'w') as outfile:
#         print(data)
        json.dump(data, outfile)
```
